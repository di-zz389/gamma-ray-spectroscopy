```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline

# Your provided data
count_rate = [138, 1294, 6416, 25585, 79671, 110181, 134548, 153659, 161815, 173501, 176742, 179653, 182590, 183593, 184010, 184734, 189635, 191991, 193392]
hv = [380, 400, 420, 440, 460,480,500, 520, 540, 560, 580, 600, 620, 640, 660,680, 700, 720, 740]

# Set figure size FIRST, before any plotting
plt.figure(figsize=(10, 5))
plt.scatter(hv, count_rate, label='Data Points')

cs = CubicSpline(hv, count_rate)
hv_smooth = np.linspace(min(hv), max(hv), 300)
plt.plot(hv_smooth, cs(hv_smooth), 'r-', label='Cubic Spline', linewidth=2)

plt.title(f'Graph of Count Rate vs HV')
plt.xlabel('HV')
plt.ylabel('Count Rate')
plt.grid(True)
plt.legend()
```
